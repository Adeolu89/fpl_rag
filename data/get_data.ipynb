```python
import pandas as pd
```

```python
eos_df = pd.read_csv("end_of_season_players.csv")
all_gw_df = pd.read_csv("all_gw.csv")
```

```python
eos_df['now_cost'] =  eos_df['now_cost']/10
```

```python
eos_df[eos_df['goals_scored'] + eos_df['assists'] >= 20] # curious about players who got 20 G+A
```

```python
gw1 = pd.read_csv(r"gws\gw1.csv")
gw1['value'] = gw1['value'] / 10
```

```python
gw1_cost = gw1[['name','value']]
```

```python
all_gw_df['value'] = all_gw_df['value'] / 10
```

```python
gw1_cost.rename(columns={"value": "gw1_cost"}, inplace=True)
```

```python
gw1_cost.info()
```

```python
eos_df["name"] = eos_df["first_name"] + " " + eos_df["second_name"]
```

```python
eos_df.head()
```

```python
eos_df = pd.merge(eos_df, gw1_cost, on='name')
```

```python
# Reorder columns in eos_df
columns_order = ['name', 'gw1_cost', 'now_cost'] + [col for col in eos_df.columns if col not in ['name', 'gw1_cost', 'now_cost']]
eos_df = eos_df[columns_order]
```

```python
eos_df.drop(columns=['first_name', 'second_name'], inplace=True)
```

```python
eos_df['price_diff'] = round(eos_df['now_cost'] -  eos_df['gw1_cost'], 2)
```

```python
eos_df.sort_values(by='price_diff', ascending=False)
```
